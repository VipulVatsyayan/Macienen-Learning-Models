## Why worry about Skewness:
### Many of the algorithms in ML work with the assumption that the data is normally distributed. So, we try to make our data as close to normal as possible. Some of the common techniques used to convert skewed data to normal distribution.

### In this case we can see that the data is Right Skewed

* Right skewed to Normal Disribution : We need to pull the tail in, because the value at the end of the tail will be affected more than the ones in the middle. So, we will use
    * Logarithm
    * Square-Root 
* For Left skewed data, we first convert them to right skew by taking the reciprocal and then applying the above techniques to make it normal    




```python
#Taking the log transform to convert the skewness and then drawing the histogram

train["SalePrice"] = np.log1p(train["SalePrice"])  

sns.distplot(train['SalePrice'])
plt.title('The skewness after log transform' + '  ' + str(train["SalePrice"].skew().round(2)))
```

# Handling missing values

```python
a= pd.DataFrame(train.isnull().sum()/ len(train))
a.columns = ['Missing Percent']
```

```python
# Sorting the values to make the graph easy to follow
a = a.sort_values(by= 'Missing Percent', ascending = False)
plt.figure(figsize = (15,5))
sns.barplot(x=a.index, y= 'Missing Percent', data =a)
plt.xticks(rotation =90)
```

We'll consider that when more than 15% of the data is missing, we should delete the corresponding variable and pretend it never existed. This means that we will not try any trick to fill the missing data in these cases. According to this, there is a set of variables (e.g. 'PoolQC', 'MiscFeature', 'Alley', etc.) that we should delete. 

```python
# Lets Check the data type

train.info()
```

```python
#LotFrontage : Since the area of each street connected to the house property most likely have a similar area
#to other houses in its neighborhood , we can fill in missing values by the median LotFrontage of the neighborhood.

train['LotFrontage'] = train['LotFrontage'].fillna(train['LotFrontage'].median())
train['MSZoning'] = train['MSZoning'].fillna(train['MSZoning'].mode()[0])
train['Electrical'] = train['Electrical'].fillna(train['Electrical'].mode()[0])
train['KitchenQual'] = train['KitchenQual'].fillna(train['KitchenQual'].mode()[0])
train['Exterior1st'] = train['Exterior1st'].fillna(train['Exterior1st'].mode()[0])
train['Exterior2nd'] = train['Exterior2nd'].fillna(train['Exterior2nd'].mode()[0])
train['SaleType'] = train['SaleType'].fillna(train['SaleType'].mode()[0])
```

```python
#Checking if the values are filled
train['LotFrontage'].isna().sum()
```

```python
# For GarageType, GarageFinish, GarageQual, GarageCond, BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF,
#BsmtFullBath and BsmtHalfBath
#we will replace missing data with 'None' as we can presume that if these items were not filled in 
#means that the house lacked those facilities
for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond','BsmtQual', 
            'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType',
            'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'] :
    train[col] = train[col].fillna('None')
```

```python
#GarageYrBlt, GarageArea and GarageCars we replace missing data with 0 (No garage = no cars in such garage.)
for col in [ 'GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 
            'BsmtHalfBath','MasVnrArea', 'MSZoning'] :
    train[col] = train[col].fillna(0)
```

```python
train = train.drop(['Alley','GarageYrBlt', 'FireplaceQu', 'PoolQC', 'Fence','MiscFeature', 'FireplaceQu', 'Utilities'], axis=1)
```

```python
#Now Checking For the skewness of other variables
numeric_feats = train.dtypes[train.dtypes != "object"].index
# Check the skew of all numerical features
skewed_feats = train[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(10)
```

```python
# Selecting data with skewness greater than .75 

skewness = skewness[abs(skewness) > 0.75]
```

```python
# Making a list of column names for the for loop
from scipy.special import boxcox1p
```

###  Earlier we did log transform because we knew it was Right skewed data. Here we dont know and it will take time so we use BoxCox1p transform to deal with that.
* Documentation on BoxCox transform can be found in the below link

https://www.statisticshowto.datasciencecentral.com/box-cox-transformation/

```python
skewed_col = skewness.index

for col in skewed_col:
    train[col] = boxcox1p(train[col], 0.15)
```

```python
train = pd.get_dummies(train)
print(train.shape)
```

```python
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Validation function
n_folds = 5
y_train = train.SalePrice.values
def rmse_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train.values)
    rmse= np.sqrt(-cross_val_score(model, train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005, random_state=1))
```

```python
rmse_cv(lasso)
```

```python
rmse_cv(lasso).mean()
```

```python
elastic_net = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9, random_state=3))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gboost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
```

```python
rmse_cv(elastic_net).mean()
```

```python
rmse_cv(gboost).mean()
```
